```python
import matplotlib.pyplot as plt

# Extended fitness tracker data for Emily Park
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
steps = [7240, 4150, 8100, 5800, 7650, 14320, 11850]
sleep_hours = [7.5, 7.0, 7.2, 4.2, 8.1, 8.5, 7.8]
stress_scores = [45, 88, 50, 75, 40, 18, 22]

# Newly included metrics
hrv = [54, 38, 56, 32, 48, 65, 62]            
deep_sleep = [1.4, 1.1, 1.3, 0.5, 1.8, 2.1, 1.6] 
calories_burned = [2150, 1850, 2300, 1980, 2200, 2950, 2600]

# Updated Dynamic VO2 Max Data from Scikit-learn output
vo2_max = [38.5, 38.5, 38.5, 38.4, 38.5, 38.9, 38.8] 

# Create a figure with a 2x3 grid of subplots
fig, axs = plt.subplots(2, 3, figsize=(16, 10))
fig.suptitle("Emily Park's 7-Day Fitness Tracker Analytics", fontsize=18, fontweight='bold', y=0.98)

# 1. Line Chart: Daily Steps
axs[0, 0].plot(days, steps, marker='o', color='royalblue', linewidth=2)
axs[0, 0].set_title('Daily Steps', fontsize=12, fontweight='bold')
axs[0, 0].set_ylabel('Steps')
axs[0, 0].grid(True, linestyle='--', alpha=0.5)

# 2. Bar Chart: Daily Sleep Hours (Total vs Deep Sleep stacked)
axs[0, 1].bar(days, sleep_hours, color='mediumaquamarine', edgecolor='black', alpha=0.7, label='Light/REM')
axs[0, 1].bar(days, deep_sleep, color='darkseagreen', edgecolor='black', label='Deep Sleep')
axs[0, 1].set_title('Daily Sleep Architecture', fontsize=12, fontweight='bold')
axs[0, 1].set_ylabel('Hours')
axs[0, 1].legend()
axs[0, 1].grid(axis='y', linestyle='--', alpha=0.5)

# 3. Line Chart: Daily Stress Score
axs[0, 2].plot(days, stress_scores, marker='s', color='crimson', linewidth=2)
axs[0, 2].set_title('Daily Stress Score', fontsize=12, fontweight='bold')
axs[0, 2].set_ylabel('Score (0-100)')
axs[0, 2].grid(True, linestyle='--', alpha=0.5)

# 4. Line Chart: Heart Rate Variability (HRV)
axs[1, 0].plot(days, hrv, marker='v', color='darkorchid', linewidth=2)
axs[1, 0].set_title('Heart Rate Variability (HRV)', fontsize=12, fontweight='bold')
axs[1, 0].set_ylabel('ms')
axs[1, 0].grid(True, linestyle='--', alpha=0.5)

# 5. Bar Chart: Calories Burned
axs[1, 1].bar(days, calories_burned, color='coral', edgecolor='black', alpha=0.8)
axs[1, 1].set_title('Calories Burned', fontsize=12, fontweight='bold')
axs[1, 1].set_ylabel('kcal')
axs[1, 1].grid(axis='y', linestyle='--', alpha=0.5)

# 6. Line Chart: Dynamic VO2 Max Estimate (Updated to match Scikit-learn output)
axs[1, 2].plot(days, vo2_max, marker='^', color='#00BCD4', linewidth=2.5, markersize=8) # Cyan matching the scikit-learn plot
axs[1, 2].set_title('Estimated Daily VO2 Max', fontsize=12, fontweight='bold')
axs[1, 2].set_ylabel('ml/kg/min')
# Setting Y-limits slightly wider than the min/max of the data to match the visual scale
axs[1, 2].set_ylim(38.35, 38.95)
axs[1, 2].grid(True, linestyle='--', alpha=0.5)

# Set common X-axis labels for the bottom row
for ax in axs[1, :]:
    ax.set_xlabel('Day of the Week')

# Optimize layout spacing
plt.tight_layout()

# Display the charts
plt.show()
```
